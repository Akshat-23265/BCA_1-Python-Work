```python
# Q1. Add names in name.txt and perform operations

with open("name.txt", "w") as f:
    n = int(input("Enter number of names: "))
    for i in range(n):
        f.write(input() + "\n")

with open("name.txt", "r") as f:
    names = [line.strip() for line in f]

print("Count of names:", len(names))

vowels = "AEIOUaeiou"
count_vowel = 0
for name in names:
    if name and name[0] in vowels:
        count_vowel += 1

print("Names starting with vowel:", count_vowel)

longest = names[0]
for name in names:
    if len(name) > len(longest):
        longest = name

print("Longest name:", longest)
```

```python
# Q2. Store integers in a file and perform operations

with open("numbers.txt", "w") as f:
    n = int(input("Enter number of integers: "))
    for i in range(n):
        f.write(input() + "\n")

with open("numbers.txt", "r") as f:
    nums = [int(line.strip()) for line in f]

max_num = nums[0]
for i in nums:
    if i > max_num:
        max_num = i

print("Maximum number:", max_num)
print("Average:", sum(nums) / len(nums))

count = 0
for i in nums:
    if i > 100:
        count += 1

print("Numbers greater than 100:", count)
```

```python
# Q3. Read city.txt and perform operations

with open("city.txt", "r") as f:
    cities = [line.strip().split() for line in f]

print("All city details:")
for c in cities:
    print(c[0], c[1], c[2])

print("Cities with population more than 10 lakhs:")
for c in cities:
    if float(c[1]) > 10:
        print(c[0])

total_area = 0
for c in cities:
    total_area += float(c[2])

print("Sum of areas:", total_area)
```

```python
# Q4. Integer division with exception handling

n = int(input("Enter number of test cases: "))

for i in range(n):
    a, b = input().split()
    try:
        print(int(a) // int(b))
    except Exception as e:
        print("Error Code:", e)
```

```python
# Q5. Multiple suitable exceptions for a file handling program

try:
    fname = input("Enter file name: ")
    f = open(fname, "r")
    data = f.read()
    print(data)
    f.close()

except FileNotFoundError:
    print("File does not exist")

except PermissionError:
    print("Permission denied")

except IsADirectoryError:
    print("Given name is a directory")

except Exception as e:
    print("Error:", e)
```
