# Experiment 3: Conditional Statements


```python
# Q1. Check whether a given number is divisible by 3 and 5 both

n = int(input("Enter a number: "))

if n % 3 == 0 and n % 5 == 0:
    print("Divisible by 3 and 5 both")
else:
    print("Not divisible by both")
```

```python
# Q2. Check whether a given number is multiple of five or not

n = int(input("Enter a number: "))

if n % 5 == 0:
    print("Multiple of 5")
else:
    print("Not a multiple of 5")
```

```python
# Q3. Find the greatest among two numbers. If numbers are equal then print "numbers are equal"

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

if a > b:
    print("Greatest:", a)
elif b > a:
    print("Greatest:", b)
else:
    print("numbers are equal")
```

```python
# Q4. Find the greatest among three numbers assuming no two values are same

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a > b and a > c:
    print("Greatest:", a)
elif b > a and b > c:
    print("Greatest:", b)
else:
    print("Greatest:", c)
```

```python
# Q5. Check whether the quadratic equation has real roots or imaginary roots. Display the roots

import math

a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

d = b*b - 4*a*c

if d > 0:
    r1 = (-b + math.sqrt(d)) / (2*a)
    r2 = (-b - math.sqrt(d)) / (2*a)
    print("Real and different roots")
    print("Roots:", r1, r2)
elif d == 0:
    r = -b / (2*a)
    print("Real and equal roots")
    print("Root:", r)
else:
    real = -b / (2*a)
    imag = math.sqrt(-d) / (2*a)
    print("Imaginary roots")
    print("Roots:", complex(real, imag), complex(real, -imag))
```

```python
# Q6. Find whether a given year is a leap year or not

year = int(input("Enter year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

```python
# Q7. Write a program which takes any date as input and display next date of the calendar

day = int(input("Enter day: "))
month = int(input("Enter month: "))
year = int(input("Enter year: "))

days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    days_in_month[1] = 29

day += 1

if day > days_in_month[month - 1]:
    day = 1
    month += 1

if month > 12:
    month = 1
    year += 1

print("Next date:", day, month, year)
```

```python
# Q8. Print the grade sheet of a student for the given range of cgpa

name = input("Enter Name: ")
roll = input("Enter Roll Number: ")
sapid = input("Enter SAPID: ")
sem = input("Enter Semester: ")
course = input("Enter Course: ")

m1 = float(input("Enter marks of Subject 1: "))
m2 = float(input("Enter marks of Subject 2: "))
m3 = float(input("Enter marks of Subject 3: "))
m4 = float(input("Enter marks of Subject 4: "))
m5 = float(input("Enter marks of Subject 5: "))

total = m1 + m2 + m3 + m4 + m5
percentage = total / 5
cgpa = percentage / 10

if cgpa <= 3.4:
    grade = "F"
elif cgpa <= 5.0:
    grade = "C+"
elif cgpa <= 6.0:
    grade = "B"
elif cgpa <= 7.0:
    grade = "B+"
elif cgpa <= 8.0:
    grade = "A"
elif cgpa <= 9.0:
    grade = "A+"
else:
    grade = "O (Outstanding)"

print("\n----- Grade Sheet -----")
print("Name:", name)
print("Roll Number:", roll, "SAPID:", sapid)
print("Sem:", sem, "Course:", course)
print("\nSubject Marks:")
print("Subject 1:", m1)
print("Subject 2:", m2)
print("Subject 3:", m3)
print("Subject 4:", m4)
print("Subject 5:", m5)
print("Percentage:", percentage, "%")
print("CGPA:", cgpa)
print("Grade:", grade)
```
