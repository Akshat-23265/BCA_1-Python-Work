```python
# Q1. Create a class of student and display details of 3 students

class Student:
    def __init__(self, name, sapid, phy, chem, maths):
        self.name = name
        self.sapid = sapid
        self.marks = [phy, chem, maths]

    def display(self):
        print("Name:", self.name)
        print("SAP ID:", self.sapid)
        print("Marks:", self.marks)

students = []

for i in range(3):
    name = input("Enter name: ")
    sapid = input("Enter SAP ID: ")
    phy = float(input("Enter Physics marks: "))
    chem = float(input("Enter Chemistry marks: "))
    maths = float(input("Enter Maths marks: "))
    students.append(Student(name, sapid, phy, chem, maths))

for s in students:
    s.display()
```

```python
# Q2. Constructor, methods and class average

class Student:
    def __init__(self, name, sapid, phy, chem, maths):
        self.name = name
        self.sapid = sapid
        self.marks = [phy, chem, maths]

    def display(self):
        print("Name:", self.name)
        print("SAP ID:", self.sapid)
        print("Marks:", self.marks)

    def marks_percentage(self):
        return sum(self.marks) / 3

    def result(self):
        for m in self.marks:
            if m < 40:
                return "Fail"
        return "Pass"

def class_average(students):
    total = 0
    for s in students:
        total += s.marks_percentage()
    return total / len(students)

n = int(input("Enter number of students: "))
students = []

for i in range(n):
    name = input("Enter name: ")
    sapid = input("Enter SAP ID: ")
    phy = float(input("Enter Physics marks: "))
    chem = float(input("Enter Chemistry marks: "))
    maths = float(input("Enter Maths marks: "))
    students.append(Student(name, sapid, phy, chem, maths))

for s in students:
    s.display()
    print("Percentage:", s.marks_percentage())
    print("Result:", s.result())

print("Class Average:", class_average(students))
```

```python
# Q3. Different types of inheritance

class A:
    def showA(self):
        print("Class A")

class B(A):
    def showB(self):
        print("Class B")

class C(A):
    def showC(self):
        print("Class C")

class D(B, C):
    def showD(self):
        print("Class D")

obj = D()
obj.showA()
obj.showB()
obj.showC()
obj.showD()
```

```python
# Q4. Method overriding

class Parent:
    def show(self):
        print("Parent class method")

class Child(Parent):
    def show(self):
        print("Child class method")

obj = Child()
obj.show()
```

```python
# Q5. Operator overloading to add two Point objects

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def display(self):
        print("x =", self.x, "y =", self.y)

p1 = Point(10, 20)
p2 = Point(12, 15)

p3 = p1 + p2
p3.display()
```
