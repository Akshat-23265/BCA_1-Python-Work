```python
#Q8- Write a program to find area of triangle when length of sides are given.
import math
a=float(input("Enter The Lenght Of Side 1 "))
b=float(input("Enter The Lenght Of Side 2 "))
c=float(input("Enter The Lenght Of Side 3 "))
s=(a+b+c)/2
area= math.sqrt(s*(s-a)*(s-b)*(s-c))
print(area)
```

```python
#Q12 - Write a program to print truth table for bitwise operators (&, |, and ^ operators)
n=[(0,0),(0,1),(1,0),(1,1)]
print("a|b|a&b|a|b|a^b")
print("-"*23)
for a,b in n:
    andd=a&b
    orr= a|b
    up=a^b
    print(andd,orr,up)
```

```python
#Q13- Write a program to find left shift and right shift values of a given number.
num = int(input("Enter a number: "))
shift = int(input("Enter shift value: "))
left = num << shift
right = num >> shift
print("Left Shift:", left)
print("Right Shift:", right)
```

```python
#Q14- Using membership operator to check number in sequence (10,20,56,78,89)
sequence = (10, 20, 56, 78, 89)
num = int(input("Enter a number: "))
if num in sequence:
    print("Number is present in sequence")
else:
    print("Number is NOT present in sequence")
```
