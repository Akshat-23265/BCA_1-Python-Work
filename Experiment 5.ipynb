# Experiment 5: String and Sets


```python
# Q1. Write a program to count and display the number of capital letters in a given string

s = input("Enter a string: ")
count = 0

for ch in s:
    if ch.isupper():
        count += 1

print("Number of capital letters:", count)
```

```python
# Q2. Count total number of vowels in a given string

s = input("Enter a string: ")
count = 0

for ch in s.lower():
    if ch in "aeiou":
        count += 1

print("Total vowels:", count)
```

```python
# Q3. Input a sentence and print words in separate lines

s = input("Enter a sentence: ")

for word in s.split():
    print(word)
```

```python
# Q4. Enter a string and a substring and print number of times substring occurs

s = input("Enter main string: ")
sub = input("Enter substring: ")

count = 0
for i in range(len(s) - len(sub) + 1):
    if s[i:i+len(sub)] == sub:
        count += 1

print(count)
```

```python
# Q5. Count the number of occurrences of each alphabet (case insensitive)

s = input("Enter a string: ").upper()
done = set()

for ch in s:
    if ch.isalpha() and ch not in done:
        print(s.count(ch), ch, sep="")
        done.add(ch)
```

```python
# Q6. Program to count the number of unique words in a given sentence using sets

s = input("Enter a sentence: ")
words = s.split()
unique_words = set(words)

print("Number of unique words:", len(unique_words))
```

```python
# Q7. Create 2 sets s1 and s2 of n fruits each and perform operations

n = int(input("Enter number of fruits in each set: "))

s1 = set()
s2 = set()

print("Enter fruits for s1:")
for i in range(n):
    s1.add(input())

print("Enter fruits for s2:")
for i in range(n):
    s2.add(input())

print("Fruits in both sets:", s1.intersection(s2))
print("Fruits only in s1:", s1.difference(s2))
print("Count of all fruits:", len(s1.union(s2)))
```

```python
# Q8. Take two sets and apply various set operations on them

s1 = {"Red", "yellow", "orange", "blue"}
s2 = {"violet", "blue", "purple"}

print("Union:", s1.union(s2))
print("Intersection:", s1.intersection(s2))
print("Difference (s1 - s2):", s1.difference(s2))
print("Difference (s2 - s1):", s2.difference(s1))
print("Symmetric Difference:", s1.symmetric_difference(s2))
```
