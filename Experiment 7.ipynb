```python
# Q1. Function to find maximum and minimum numbers from a sequence (without built-in functions)

def find_max_min(seq):
    max_val = seq[0]
    min_val = seq[0]

    for i in seq:
        if i > max_val:
            max_val = i
        if i < min_val:
            min_val = i

    return max_val, min_val

nums = list(map(int, input("Enter numbers: ").split()))
print(find_max_min(nums))
```

```python
# Q2. Function that returns sum of cubes of all positive integers smaller than given number

def sum_of_cubes(n):
    s = 0
    for i in range(1, n):
        s += i ** 3
    return s

n = int(input("Enter a number: "))
print(sum_of_cubes(n))
```

```python
# Q3. Function to print 1 to n using recursion (without loop)

def print_numbers(n, i=1):
    if i > n:
        return
    print(i)
    print_numbers(n, i + 1)

n = int(input("Enter n: "))
print_numbers(n)
```

```python
# Q4. Recursive function to print Fibonacci series upto n terms

def fib(a, b, n):
    if n == 0:
        return
    print(a, end=" ")
    fib(b, a + b, n - 1)

n = int(input("Enter number of terms: "))
fib(0, 1, n)
```

```python
# Q5. Lambda function to find volume of cone

volume_cone = lambda r, h: (3.14 * r * r * h) / 3

r = float(input("Enter radius: "))
h = float(input("Enter height: "))

print("Volume of cone:", volume_cone(r, h))
```

```python
# Q6. Lambda function which gives tuple of max and min from a list

lst = list(map(int, input("Enter numbers: ").split()))

max_min = lambda l: (max(l), min(l))

print(max_min(lst))
```

```python
# Q7. Functions to explain keyword argument, default argument, variable length argument

def student(name, age):
    print("Name:", name)
    print("Age:", age)

def greet(name="Student"):
    print("Hello", name)

def total(*numbers):
    s = 0
    for i in numbers:
        s += i
    print("Sum:", s)

print("Keyword argument:")
student(age=20, name="Rohit")

print("Default argument:")
greet()
greet("Akshat")

print("Variable length argument:")
total(10, 20, 30)
```
